```python
def remove_outliers_iqr(df):
    df_clean = df.copy()
    numeric_cols = df_clean.select_dtypes(include=['float64', 'int64']).columns

    for col in numeric_cols:
        Q1 = df_clean[col].quantile(0.25)
        Q3 = df_clean[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df_clean = df_clean[(df_clean[col] >= lower_bound) & (df_clean[col] <= upper_bound)]

    print("✅ Outlier removal complete!")
    print(f"Rows before: {df.shape[0]} → after: {df_clean.shape[0]}")
    return df_clean
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# === Step 1: Detect numeric columns first ===
numeric_cols = auto.select_dtypes(include=['float64', 'int64']).columns

# === Step 2: Boxplots before cleaning ===
print("📊 Boxplots before removing outliers:")

plt.figure(figsize=(15, 8))
auto[numeric_cols].boxplot()
plt.title("Before Removing Outliers")
plt.show()

# === Step 3: Define the outlier removal function ===
def remove_outliers_iqr(df):
    df_clean = df.copy()
    numeric_cols_local = df_clean.select_dtypes(include=['float64', 'int64']).columns  # local copy inside function

    for col in numeric_cols_local:
        Q1 = df_clean[col].quantile(0.25)
        Q3 = df_clean[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df_clean = df_clean[(df_clean[col] >= lower_bound) & (df_clean[col] <= upper_bound)]

    print("✅ Outlier removal complete!")
    print(f"Rows before: {df.shape[0]} → after: {df_clean.shape[0]}")
    return df_clean

# === Step 4: Apply outlier removal ===
auto_cleaned = remove_outliers_iqr(auto)

# === Step 5: Boxplots after cleaning ===
print("\n📊 Boxplots after removing outliers:")

plt.figure(figsize=(15, 8))
auto_cleaned[numeric_cols].boxplot()
plt.title("After Removing Outliers")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression #sklearn.linear_model will support LR & logistic
from sklearn.model_selection import train_test_split
```

```python
x=auto[auto.columns[:-1]]
y=auto['price']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, train_size =0.75, random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
l_model=LinearRegression()
```

```python
l_model.fit(x_train,y_train)
```

```python
y_pred =l_model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import *
```

```python
r2_on_test = r2_score(y_test,y_pred) #how far the model have learnt--> r2_score
r2_on_test
```

```python
weights= l_model.coef_
intercept= l_model.intercept_

print("Weights:", weights)
print("Intercept:", intercept)
```

```python
from sklearn.tree import *
model = DecisionTreeRegressor(random_state=24, criterion='squared_error', splitter='best', max_depth=None)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
import math # no dect key for rmse so we are doing it
r2 = r2_score(y_test, y_pred)
mse=mean_squared_error(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
rmse=math.sqrt(mean_squared_error(y_test,y_pred))
print(mse,mae,rmse)
print(f"Model Accuracy: {r2*100:.2f}%")
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor

param_grid = {
    'max_depth': [3, 5, 7, 10, None],
    'min_samples_split': [2, 5, 10, 20],
    'min_samples_leaf': [1, 2, 5, 10],
    'criterion': ['squared_error', 'absolute_error'],
    'splitter': ['best', 'random']
}

grid = GridSearchCV(
    DecisionTreeRegressor(random_state=42),
    param_grid,
    scoring='r2',
    cv=5
)
grid.fit(x_train, y_train)

print("Best Parameters:", grid.best_params_)
print("Best R²:", grid.best_score_)
```
